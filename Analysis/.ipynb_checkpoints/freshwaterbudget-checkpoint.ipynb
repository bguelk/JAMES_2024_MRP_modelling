```python
# import necessary libraries for this calculation
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cmocean
from pathlib import Path
import scipy

plt.rcParams['font.size']=14
```

# REFERENCE EXPERIMENT

```python
ds01=xr.open_mfdataset('./EXP_REF_intel/fw_20*_EXP_REF_intel.nc')

time_array=np.array(ds01.t,dtype=np.datetime64)
```

```python
S0=35
dt=86400
```

```python
rd=30
plt.figure(figsize=(8,6))
(ds01.fw_adv_wb*dt).rolling(t=rd,center=True).mean().plot(color='k',linestyle='-',label='WB')
(ds01.fw_adv_nb*dt).rolling(t=rd,center=True).mean().plot(color='b',linestyle='-',label='NB')
(ds01.fw_adv_eb*dt).rolling(t=rd,center=True).mean().plot(color='r',linestyle='-',label='EB')
plt.grid()
plt.legend()
```

```python
dt=86400
dFc_dt01=((ds01.Fc.diff(dim='t'))/dt).isel(t=slice(1,None,None))
fluxes01=((-1*ds01.Fsurf+(ds01.fw_adv_eb+ds01.fw_adv_nb+ds01.fw_adv_wb)).rolling(t=2,center=True).mean()).isel(t=slice(2,None,None))

dFc_dt01.load()
```

```python
rd=30
fig,axes=plt.subplots(5,1,figsize=(8,10))
ax=axes[0]
((-1*ds01.Fsurf)).rolling(t=rd,center=True).mean().plot(ax=ax,color='y',linestyle='-',label='Fsurf')
ax.grid()
ax.set_xticklabels('')
ax.set_xlabel('')
ax.set_ylabel('surface flux')
ax.legend()
ax=axes[1]
ds01.fw_adv_eb.rolling(t=rd,center=True).mean().plot(ax=ax,color='g',linestyle='-',label='EB')
ds01.fw_adv_nb.rolling(t=rd,center=True).mean().plot(ax=ax,color='r',linestyle='-',label='NB')
ds01.fw_adv_wb.rolling(t=rd,center=True).mean().plot(ax=ax,color='b',linestyle='-',label='WB')

ax.grid()
ax.set_xticklabels('')
ax.set_xlabel('')
ax.set_ylabel('lateral fluxes')
ax.legend()
ax=axes[2]
dFc_dt01.isel(t=slice(1,None,None)).rolling(t=rd,center=True).mean().plot(ax=ax,color='k',linestyle='-',label='Fc')
ax.set_ylabel('dFc /dt')
ax.set_xticklabels('')
ax.set_xlabel('')
#ax.set_ylim(6.15e13,6.30e13)
ax.grid()

ax=axes[3]
(dFc_dt01-fluxes01).isel(t=slice(1,None,None)).rolling(t=rd,center=True).mean().plot(ax=ax,color='k',linestyle='-',label='Res')
ax.set_ylabel('Residual')
ax.set_xticklabels('')
ax.set_xlabel('')
#ax.set_ylim(6.15e13,6.30e13)
ax.grid()

ax=axes[4]
ds01.Fc.rolling(t=rd,center=True).mean().plot(ax=ax,color='r',linestyle='-',label='Fc')

ax.set_ylabel('$F_{content}$')
ax.set_xlabel('')

ax.grid()
plt.tight_layout()
```

# NOW include the other full runs 

```python
ds03=xr.open_mfdataset('./EXP_mfc_ori/fw_20*_EXP_mfc_ori.nc')
ds04=xr.open_mfdataset('./EXP_mfc_mod/fw_20*_EXP_mfc_mod.nc')
```

```python
# focus on F content

ds01.Fc.plot(color='k',linestyle='-',label='EVD')
ds03.Fc.plot(color='r',linestyle='-',label='EDMF')
ds04.Fc.plot(color='b',linestyle='-',label='$EDMF_{\\epsilon}$')
plt.grid()
plt.legend()
plt.ylabel('$F_c$ [$m^3$]')
plt.xlabel('')
plt.xlim(time_array[1],time_array[-1])
plt.savefig('./figures/freshwatercontent_fullruns.jpg',dpi=300)
```

```python
# check if trend is found by lin reg
res01= scipy.stats.linregress(np.arange(0,len(time_array)), ds01.Fc, alternative='two-sided')
res03= scipy.stats.linregress(np.arange(0,len(time_array)), ds03.Fc, alternative='two-sided')
res04= scipy.stats.linregress(np.arange(0,len(time_array)), ds04.Fc, alternative='two-sided')
ds01.Fc.load().plot(color='k',linestyle=':',label='REF')
plt.plot(time_array, res01.intercept + res01.slope*np.arange(0,len(time_array)), 'k')#, label='fitted line')
ds03.Fc.load().plot(color='r',linestyle=':',label='EDMF')
plt.plot(time_array, res03.intercept + res03.slope*np.arange(0,len(time_array)), 'r')#, label='fitted line')
ds04.Fc.load().plot(color='b',linestyle=':',label='EDMF C3')
plt.plot(time_array, res04.intercept + res04.slope*np.arange(0,len(time_array)), 'b')#, label='fitted line')
plt.grid()
plt.legend()
plt.xlim(time_array[0],time_array[-1])
```

```python
ds01.Fc.plot(color='k',linestyle='-',label='original data')
ds01.Fc.load().rolling(t=rd,center=True).mean().plot(color='grey',linestyle='-',label='1 year running mean')
plt.plot(time_array, res01.intercept + res01.slope*np.arange(0,len(time_array)), 'r', label='fitted line')
plt.legend()
plt.grid()
```

```python
print(res01.intercept)
print(res03.intercept)
print(res04.intercept)
```

```python
print(res01.slope/(86400*(10**3)),'',res01.stderr/(86400*(10**3)),'$10^{-3}$ Sv')
print(res03.slope/(86400*(10**3)),'',res03.stderr/(86400*(10**3)),'$10^{-3}$ Sv')
print(res04.slope/(86400*(10**3)),'',res04.stderr/(86400*(10**3)),'$10^{-3}$ Sv')
```

# NOW sensitivity runs - Precip

```python
ds34=xr.open_mfdataset('./EXP_initial/fw_20*_EXP_initial.nc')
ds36=xr.open_mfdataset('./EXP_tpre00_prsn00_2012/fw_201*_EXP_tpre00_prsn00_2012.nc')
ds38=xr.open_mfdataset('./EXP_tpre01_prsn01_2012/fw_201*_EXP_tpre01_prsn01_2012.nc')
ds39=xr.open_mfdataset('./EXP_tpre02_prsn02_2012/fw_201*_EXP_tpre02_prsn02_2012.nc')
ds40=xr.open_mfdataset('./EXP_tpre03_prsn03_2012/fw_201*_EXP_tpre03_prsn03_2012.nc')
ds41=xr.open_mfdataset('./EXP_tpre025_prsn025_2012/fw_201*_EXP_tpre025_prsn025_2012.nc')
ds43=xr.open_mfdataset('./EXP_tpre04_prsn04_2012/fw_201*_EXP_tpre04_prsn04_2012.nc')
```

```python
plt.figure(figsize=(10,10))
ds01.Fc.load().rolling(t=rd,center=True).mean().plot(color='k',linestyle=':',label='REF')
ds04.Fc.load().rolling(t=rd,center=True).mean().plot(color='k',linestyle='-',label='EDMF Case 3')
ds36.Fc.load().rolling(t=rd,center=True).mean().plot(label='0%')
ds38.Fc.load().rolling(t=rd,center=True).mean().plot(label='10%')
ds39.Fc.load().rolling(t=rd,center=True).mean().plot(label='20%')
ds40.Fc.load().rolling(t=rd,center=True).mean().plot(label='30%')
ds41.Fc.load().rolling(t=rd,center=True).mean().plot(label='25%')
ds43.Fc.load().rolling(t=rd,center=True).mean().plot(label='40%')
plt.xlim(time_array[1826],time_array[-1])
plt.grid()
plt.legend(ncol=3)
plt.ylabel('$F_c$ [$m^3$]')
plt.xlabel('')
```

```python
ds04.Fc.plot(color='k',linestyle='-',label='EDMF Case 3')
xr.concat((ds34.Fc,ds36.Fc),dim='t').plot(label='20%')
```

```python
# linear regression of those runs
res36= scipy.stats.linregress(np.arange(0,len(time_array)), xr.concat((ds34.Fc,ds36.Fc),dim='t'), alternative='two-sided')
res38= scipy.stats.linregress(np.arange(0,len(time_array)), xr.concat((ds34.Fc,ds38.Fc),dim='t'), alternative='two-sided')
res39= scipy.stats.linregress(np.arange(0,len(time_array)), xr.concat((ds34.Fc,ds39.Fc),dim='t'), alternative='two-sided')
res40= scipy.stats.linregress(np.arange(0,len(time_array)), xr.concat((ds34.Fc,ds40.Fc),dim='t'), alternative='two-sided')
res41= scipy.stats.linregress(np.arange(0,len(time_array)), xr.concat((ds34.Fc,ds41.Fc),dim='t'), alternative='two-sided')
res43= scipy.stats.linregress(np.arange(0,len(time_array)), xr.concat((ds34.Fc,ds43.Fc),dim='t'), alternative='two-sided')


plt.plot(time_array, res01.intercept + res01.slope*np.arange(0,len(time_array)), 'k', label='REF',ls='--')
plt.plot(time_array, res03.intercept + res03.slope*np.arange(0,len(time_array)), 'r', label='EDMF',ls='--')
ds04.Fc.load().plot(color='grey',linestyle=':')#,label='EDMF C3')
plt.plot(time_array, res04.intercept + res04.slope*np.arange(0,len(time_array)), 'b', label='EDMF C3',ls='--')
plt.plot(time_array, res36.intercept + res36.slope*np.arange(0,len(time_array)),label='0%')
plt.plot(time_array, res38.intercept + res38.slope*np.arange(0,len(time_array)),label='10%')
plt.plot(time_array, res39.intercept + res39.slope*np.arange(0,len(time_array)),label='20%')
plt.plot(time_array, res40.intercept + res40.slope*np.arange(0,len(time_array)),label='30%')
plt.plot(time_array, res41.intercept + res41.slope*np.arange(0,len(time_array)),label='25%')
plt.plot(time_array, res43.intercept + res43.slope*np.arange(0,len(time_array)),label='40%')


plt.grid()
plt.legend(ncol=3)
plt.ylabel('$F_c$ [$m^3$]')
plt.xlabel('')
plt.xlim(time_array[0],time_array[-1])
```

```python
print('REF:    ',np.round(res01.slope/(86400*(10**3)),2),'',np.round(res01.stderr/(86400*(10**3)),2),'$10^{-3}$ Sv')
print('EDMF:   ',np.round(res03.slope/(86400*(10**3)),2),'',np.round(res03.stderr/(86400*(10**3)),2),'$10^{-3}$ Sv')
print('EDMF C3:',np.round(res04.slope/(86400*(10**3)),2),'',np.round(res04.stderr/(86400*(10**3)),2),'$10^{-3}$ Sv')

print('0% :    ', np.round(res36.slope/(86400*(10**3)),2),'',np.round(res36.stderr/(86400*(10**3)),2),'$10^{-3}$ Sv')
print('10%:    ', np.round(res38.slope/(86400*(10**3)),2),'',np.round(res38.stderr/(86400*(10**3)),2),'$10^{-3}$ Sv')
print('20%:    ', np.round(res39.slope/(86400*(10**3)),2),'',np.round(res39.stderr/(86400*(10**3)),2),'$10^{-3}$ Sv')
print('30%:    ', np.round(res40.slope/(86400*(10**3)),2),'',np.round(res40.stderr/(86400*(10**3)),2),'$10^{-3}$ Sv')
print('25%:    ', np.round(res41.slope/(86400*(10**3)),2),'',np.round(res41.stderr/(86400*(10**3)),2),'$10^{-3}$ Sv')
print('40%:    ', np.round(res43.slope/(86400*(10**3)),2),'',np.round(res43.stderr/(86400*(10**3)),2),'$10^{-3}$ Sv')
```
