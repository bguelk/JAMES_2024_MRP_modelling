```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import gsw
import cmocean.cm as cmo
#import cf_xarray as cfxr
import xarray as xr
from skimage.morphology import dilation
from skimage.morphology import disk
import datetime
xr.set_options(keep_attrs=True)
plt.rcParams['font.size']=14
```

```python
dswp16=xr.open_mfdataset('./EXP_tpre02_prsn02_2012/wp_dcreg_2016_tpre02_prsn02_2012.nc')
dswp17=xr.open_mfdataset('./EXP_tpre02_prsn02_2012/wp_dcreg_2017_tpre02_prsn02_2012.nc')
dsm=xr.open_mfdataset('./EXP_tpre02_prsn02_2012/mld_201[6,7].nc')
dsm.coords['year']=dsm.time_counter.dt.year
dsm=dsm.rename({'nav_lon_grid_T':'nav_lon','nav_lat_grid_T':'nav_lat'})
dshi=xr.open_mfdataset('./EXP_tpre02_prsn02_2012/hi800_prepolynya_tpre02_prsn02_2012_201*.nc')
dshi=dshi.rename({'nav_lon_grid_T':'nav_lon','nav_lat_grid_T':'nav_lat'})
dssa=xr.open_mfdataset('./EXP_tpre02_prsn02_2012/*100_saltadv_tpre02_prsn02_2012.nc')
```

```python
bat=xr.open_dataset('../MAUD_RISE_config/maud_v4/m12_mask/1_domain_cfg_50levels_new.nc')
bat.coords['lon']=bat.nav_lon
bat.coords['lat']=bat.nav_lat
```

```python
time_array16=np.array(dswp16.time_counter,dtype=np.datetime64)
time_array17=np.array(dswp17.time_counter,dtype=np.datetime64)
```

```python
years=[2016,2017]
minor_ticks = []
for y in years:
    minor_ticks += [datetime.datetime(y,m,1) for m in range(1,13)]
```

```python
dshi.hi.where(dshi.year==2017).mean(dim='t').min().values
```

```python
#from matplotlib import colors as c
dt=30
fig,axes=plt.subplots(3,3,figsize=(10,10),gridspec_kw={'width_ratios':[1,1,.1],'hspace':0.3})
# plume velocity
ax=axes[0,0]
cwp=dswp16.mf_wp.max(dim={'x','y'}).plot.pcolormesh(x='time_counter',y='depthw',ax=ax,cmap=cmo.matter,vmin=0., vmax=.04,levels=21, add_colorbar=False)
dsm.mldr10_1.where((dsm.nav_lon>1.5)&(dsm.nav_lon<5)&(dsm.nav_lat<-63.9)&(dsm.nav_lat>-64.3)).max(dim={'x','y'}).plot(x="time_counter",color='grey',ax=ax)
ax.set_ylim(1000,0)
ax.set_ylabel('Depth [m]')
ax.set_xlabel('Month')
ax.set_title('2016')
ax.vlines(x=time_array16[198-dt],ymin=0,ymax=1000,color='k',ls=':')
ax.vlines(x=time_array16[198],ymin=0,ymax=1000,color='w',ls=':')
ax.set_xticks([datetime.datetime(2016,m,1) for m in range(2,13,2)])
ax.set_xticklabels(np.arange(2,13,2))
ax.xaxis.set_tick_params(rotation=30)
ax.set_xticks(minor_ticks, minor=True);
ax.set_xlim(time_array16[0],time_array16[-1])
ax.text(0.03, 1.05, 'a)', horizontalalignment='center',verticalalignment='center', transform=ax.transAxes,fontsize='14')

#isel(t=slice(198-dt,198))
ax=axes[0,1]
dswp17.mf_wp.max(dim={'x','y'}).plot.pcolormesh(x='time_counter',y='depthw',ax=ax,cmap=cmo.matter,vmin=0., vmax=.04,levels=21, add_colorbar=False)
dsm.mldr10_1.where((dsm.nav_lon>5.3)&(dsm.nav_lon<5.9)&(dsm.nav_lat<-64.2)&(dsm.nav_lat>-64.6)).max(dim={'x','y'}).plot(x="time_counter",color='grey',ax=ax)
ax.set_ylim(1000,0)
ax.set_ylabel('')
ax.set_yticklabels('')
ax.set_xlabel('Month')
ax.set_title('2017')
plt.colorbar(cwp,cax=axes[0,2],label='Plume Velocity [$m\,s^{-1}$]',ticks=np.linspace(0,0.04,5))
ax.vlines(x=time_array17[180],ymin=0,ymax=1000,color='w',ls=':')
ax.vlines(x=time_array17[180-dt],ymin=0,ymax=1000,color='k',ls=':')
ax.set_xticks([datetime.datetime(2017,m,1) for m  in range(2,13,2)])
ax.set_xticklabels(np.arange(2,13,2))
#ax.xaxis.set_tick_params(rotation=0)
ax.set_xticks(minor_ticks, minor=True);
ax.set_xlim(time_array17[0],time_array17[-1])
ax.text(0.03, 1.05, 'b)', horizontalalignment='center',verticalalignment='center', transform=ax.transAxes,fontsize='14')

# salt deficit
ax=axes[1,0]
dshi.hi.where(dshi.year==2016).mean(dim='t').plot.contourf(x="nav_lon",y="nav_lat",vmin=0,vmax=1.5, levels=16,ax=ax, cmap=cmo.tempo,add_colorbar=False)
bat.bathy_meter.where(bat.bathy_meter<10).plot.contourf(x="lon",y="lat",colors='grey',add_colorbar=False,ax=ax,alpha=0.5)
bat.bathy_meter.plot.contour(x="lon",y="lat",levels=[2500,3500,5000],colors='grey',ax=ax,alpha=0.5)
ax.set_xlim(-2,10)
ax.set_ylim(-66,-63)
ax.set_xticklabels('')
ax.set_xlabel('')
ax.set_ylabel('Latitude [$^\circ$N]')
# add box of mean
ax.plot((1.5,1.5),(-64.3,-63.9),color='darkviolet')
ax.plot((5,5),(-64.3,-63.9),color='darkviolet')
ax.plot((1.5,5),(-63.9,-63.9),color='darkviolet')
ax.plot((1.5,5),(-64.3,-64.3),color='darkviolet')
ax.text(0.03, 1.05, 'c)', horizontalalignment='center',verticalalignment='center', transform=ax.transAxes,fontsize='14')

ax=axes[1,1]
chi=dshi.hi.where(dshi.year==2017).mean(dim='t').where(dshi.hi.where(dshi.year==2017).mean(dim='t')>0).plot.contourf(x="nav_lon",y="nav_lat",vmin=0,vmax=1.5, levels=16,cmap=cmo.tempo,ax=ax , add_colorbar=False)
bat.bathy_meter.where(bat.bathy_meter<10).plot.contourf(x="lon",y="lat",colors='grey',add_colorbar=False,ax=ax,alpha=0.5)
bat.bathy_meter.plot.contour(x="lon",y="lat",levels=[2500,3500,5000],colors='grey',ax=ax,alpha=0.5)
ax.set_ylabel('')
ax.set_yticklabels('')
ax.set_xticklabels('')
ax.set_xlim(-2,10)
ax.set_ylim(-66,-63)
ax.set_xlabel('')
ax.plot((5.3,5.3),(-64.6,-64.2),color='darkviolet')
ax.plot((5.9,5.9),(-64.6,-64.2),color='darkviolet')
ax.plot((5.3,5.9),(-64.2,-64.2),color='darkviolet')
ax.plot((5.3,5.9),(-64.6,-64.6),color='darkviolet')

plt.colorbar(chi,cax=axes[1,2],label='Salt Deficit [$m$]',ticks=np.linspace(0,1.5,4))
ax.text(0.03, 1.05, 'd)', horizontalalignment='center',verticalalignment='center', transform=ax.transAxes,fontsize='14')

#salt anomaly 
ax=axes[2,0]

ar=6
dssa.sa_100.where(dssa.year==2016,drop=True).isel(t=slice(198-dt,198)).mean(dim={'t'}).plot.contourf(ax=ax,x="nav_lon_grid_T",y="nav_lat_grid_T",levels=21,vmin=-0.5,vmax=0.5,cmap=cmo.balance,add_colorbar=False)
q16=dssa.where(dssa.year==2016,drop=True).isel(t=slice(198-dt,198)).mean(dim='t').isel(x_f=slice(1,-1,ar),y_c=slice(1,-1,ar)).plot.quiver(x="glamu",y="gphiu",u="u_100",v="v_100",ax=ax,add_guide=False)
#ax.quiverkey(q16,1,1,0.5,'1m/s',color='r')
bat.bathy_meter.where(bat.bathy_meter<10).plot.contourf(x="lon",y="lat",colors='grey',add_colorbar=False,ax=ax,alpha=0.5)
bat.bathy_meter.plot.contour(x="lon",y="lat",levels=[2500,3500,5000],colors='grey',ax=ax,alpha=0.5)
ax.set_ylabel('Latitude [$^\circ$N]')
ax.set_xlabel('Longitude [$^\circ$E]')
ax.set_xlim(-2,10)
ax.set_ylim(-66,-63)
ax.quiverkey(q16,0.85,0.1,0.2,'0.2m/s',color='b')
# add box of mean
ax.plot((1.5,1.5),(-64.3,-63.9),color='darkviolet')
ax.plot((5,5),(-64.3,-63.9),color='darkviolet')
ax.plot((1.5,5),(-63.9,-63.9),color='darkviolet')
ax.plot((1.5,5),(-64.3,-64.3),color='darkviolet')
ax.text(0.03, 1.05, 'e)', horizontalalignment='center',verticalalignment='center', transform=ax.transAxes,fontsize='14')

ax=axes[2,1]
csa=dssa.sa_100.where(dssa.year==2017,drop=True).isel(t=slice(180-dt,180)).mean(dim={'t'}).plot.contourf(ax=ax,x="nav_lon_grid_T",y="nav_lat_grid_T",levels=21,vmin=-.5,vmax=0.5,cmap=cmo.balance,add_colorbar=False)
q17=dssa.where(dssa.year==2017,drop=True).isel(t=slice(180-dt,180)).mean(dim='t').isel(x_f=slice(1,-1,ar),y_c=slice(1,-1,ar)).plot.quiver(x="glamu",y="gphiu",u="u_100",v="v_100",ax=ax,add_guide=False)
#plt.quiverkey(q17,8,-66.5,0.5,'0.5m/s',color='w',ax=ax)
bat.bathy_meter.where(bat.bathy_meter<10).plot.contourf(x="lon",y="lat",colors='grey',add_colorbar=False,ax=ax,alpha=0.5)
bat.bathy_meter.plot.contour(x="lon",y="lat",levels=[2500,3500,5000],colors='grey',ax=ax,alpha=0.5)
ax.set_ylabel('')
ax.set_xlabel('Longitude [$^\circ$E]')
ax.set_xlim(-2,10)
ax.set_ylim(-66,-63)
ax.set_yticklabels('')
ax.quiverkey(q17,0.85,.1,0.2,'0.2m/s',color='b')
plt.colorbar(csa,cax=axes[2,2],label='Salt Anomaly [$g\,kg^{-1}$]',ticks=np.linspace(-.5,.5,3))
# add box of mean
ax.plot((5.3,5.3),(-64.6,-64.2),color='darkviolet')
ax.plot((5.9,5.9),(-64.6,-64.2),color='darkviolet')
ax.plot((5.3,5.9),(-64.2,-64.2),color='darkviolet')
ax.plot((5.3,5.9),(-64.6,-64.6),color='darkviolet')

ax.text(0.03, 1.05, 'f)', horizontalalignment='center',verticalalignment='center', transform=ax.transAxes,fontsize='14')
plt.tight_layout()
plt.savefig('./EXP_tpre02_prsn02_2012/figures/processes_polynya.jpg',dpi=300)
```
